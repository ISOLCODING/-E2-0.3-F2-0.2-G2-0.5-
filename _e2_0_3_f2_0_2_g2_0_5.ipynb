```python
import openpyxl
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side
from openpyxl.formatting.rule import ColorScaleRule, Rule, FormatObject
from openpyxl.styles.differential import DifferentialStyle


def create_youtube_strategy_spreadsheet(filename="Strategi_Channel_YouTube.xlsx"):
    """
    Membuat file Excel lengkap untuk strategi dan manajemen channel YouTube.
    """
    wb = openpyxl.Workbook()

    # Hapus sheet default
    if "Sheet" in wb.sheetnames:
        wb.remove(wb["Sheet"])

    # === TAB 1: DASHBOARD & PERENCANAAN KONTEN ===
    ws1 = wb.create_sheet("Dashboard & Perencanaan Konten")

    # Definisikan Header dan Data
    headers_ws1 = [
        "Judul Konten", "Kategori", "Level", "Target Audiens (Spesifik)",
        "Trend (1-10)", "Kesulitan (1-10)", "Permintaan Pasar (1-10)",
        "Prioritas", "Status", "Deadline Upload"
    ]

    data_ws1 = [
        # Programming
        ("Belajar Python dari NOL dalam 15 Menit", "Programming", "Pemula", "Pelajar SMA/Mahasiswa Baru", 9, 3, 10, None, "Ide", "Minggu 1"),
        ("Setup Awal JavaScript untuk Pemula", "Programming", "Pemula", "Calon Web Developer", 8, 2, 9, None, "Ide", "Minggu 1"),
        ("Perintah Dasar GIT yang Wajib Dikuasai", "Programming", "Pemula", "Semua Programmer", 9, 4, 10, None, "Ide", "Minggu 2"),
        ("Membuat Website Portofolio dengan HTML & CSS", "Programming", "Pemula", "Mahasiswa & Job Seeker", 8, 5, 9, None, "Ide", "Minggu 3"),
        ("Memahami Konsep API untuk Pemula", "Programming", "Menengah", "Mahasiswa IT", 9, 6, 8, None, "Ide", "Minggu 5"),
        ("Project CRUD Sederhana dengan JavaScript", "Programming", "Menengah", "Web Developer Junior", 8, 7, 8, None, "Ide", "Minggu 6"),
        ("Pengenalan React JS: Kapan & Mengapa?", "Programming", "Menengah", "Programmer >1 tahun", 10, 6, 9, None, "Ide", "Minggu 7"),
        ("Mengelola State di React dengan Hooks", "Programming", "Menengah", "React Developer", 10, 7, 8, None, "Ide", "Minggu 9"),
        ("Otomatisasi Tugas Harian dengan Python", "Programming", "Lanjut", "Profesional & Mahasiswa", 9, 8, 7, None, "Ide", "Minggu 11"),
        ("Optimasi Performa Website React", "Programming", "Lanjut", "Web Developer Senior", 8, 8, 7, None, "Ide", "Minggu 12"),
        # Aplikasi
        ("10 Rumus Excel Paling Berguna di Dunia Kerja", "Aplikasi", "Pemula", "Karyawan & Admin", 9, 3, 10, None, "Ide", "Minggu 2"),
        ("Menguasai Pivot Table Excel dalam 10 Menit", "Aplikasi", "Pemula", "Analis & Karyawan", 9, 4, 9, None, "Ide", "Minggu 4"),
        ("Tips & Trik Microsoft Office 365 (Word, PPT)", "Aplikasi", "Pemula", "Pelajar & Karyawan", 8, 2, 8, None, "Ide", "Minggu 5"),
        ("Membuat Dashboard Interaktif di Excel", "Aplikasi", "Menengah", "Manajer & Analis Data", 9, 7, 9, None, "Ide", "Minggu 6"),
        ("Menghapus Background Foto di Photoshop", "Aplikasi", "Menengah", "Desainer & Konten Kreator", 8, 4, 9, None, "Ide", "Minggu 8"),
        ("Kolaborasi Efektif dengan Office 365", "Aplikasi", "Menengah", "Tim & Profesional", 7, 3, 7, None, "Ide", "Minggu 9"),
        ("Formula VLOOKUP vs INDEX MATCH Excel", "Aplikasi", "Lanjut", "Analis Data & Akuntan", 8, 7, 8, None, "Ide", "Minggu 10"),
        ("Automasi Laporan Excel dengan Macro", "Aplikasi", "Lanjut", "Staf Keuangan & Analis", 7, 9, 7, None, "Ide", "Minggu 12"),
        # Teknologi
        ("Cara Mempercepat Kinerja Windows 11/10", "Teknologi", "Pemula", "Semua Pengguna PC", 10, 3, 10, None, "Ide", "Minggu 3"),
        ("5 Ekstensi Chrome yang Bikin Produktif", "Teknologi", "Pemula", "Pelajar & Pekerja Remote", 9, 2, 9, None, "Ide", "Minggu 4"),
        ("Keyboard Shortcut Windows Wajib Tahu", "Teknologi", "Pemula", "Semua Pengguna PC", 8, 2, 8, None, "Ide", "Minggu 7"),
        ("Mengamankan Akun Google dari Hacker", "Teknologi", "Menengah", "Semua Pengguna Internet", 9, 4, 10, None, "Ide", "Minggu 8"),
        ("Memilih Laptop Terbaik untuk Programming", "Teknologi", "Menengah", "Mahasiswa IT", 8, 5, 9, None, "Ide", "Minggu 10"),
        ("Cara Membersihkan Laptop dari 'Sampah'", "Teknologi", "Menengah", "Semua Pengguna Laptop", 9, 3, 9, None, "Ide", "Minggu 11"),
    ]

    # Styling
    header_font = Font(bold=True, color="FFFFFF")
    header_fill = PatternFill(start_color="4F81BD", end_color="4F81BD", fill_type="solid")

    # Tulis header
    ws1.append(headers_ws1)
    for cell in ws1[1]:
        cell.font = header_font
        cell.fill = header_fill

    # Tulis data dan formula
    for i, row_data in enumerate(data_ws1, start=2):
        ws1.append(row_data)
        # Formula Prioritas di kolom H (indeks 8)
        formula_cell = ws1.cell(row=i, column=8)
        formula_cell.value = f"= (E{i}*0.3) + (F{i}*0.2) + (G{i}*0.5)"
        formula_cell.number_format = '0.0'

    # Conditional Formatting untuk Prioritas (Kolom H)
    red_fill = PatternFill(start_color="FFC7CE", end_color="FFC7CE", fill_type="solid")
    yellow_fill = PatternFill(start_color="FFEB9C", end_color="FFEB9C", fill_type="solid")
    green_fill = PatternFill(start_color="C6EFCE", end_color="C6EFCE", fill_type="solid")

    red_dxf = DifferentialStyle(fill=red_fill)
    yellow_dxf = DifferentialStyle(fill=yellow_fill)
    green_dxf = DifferentialStyle(fill=green_fill)

    ws1.conditional_formatting.add(f"H2:H{len(data_ws1)+1}", Rule(type="cellIs", operator="lessThan", formula=[4], dxf=red_dxf))
    ws1.conditional_formatting.add(f"H2:H{len(data_ws1)+1}", Rule(type="cellIs", operator="between", formula=[4, 7], dxf=yellow_dxf))
    ws1.conditional_formatting.add(f"H2:H{len(data_ws1)+1}", Rule(type="cellIs", operator="greaterThan", formula=[7], dxf=green_dxf))

    # Atur lebar kolom
    column_widths = {'A': 45, 'B': 15, 'C': 12, 'D': 30, 'E': 15, 'F': 15, 'G': 20, 'H': 10, 'I': 12, 'J': 15}
    for col, width in column_widths.items():
        ws1.column_dimensions[col].width = width

    # === TAB 2: KALENDER POSTING STRATEGIS ===
    ws2 = wb.create_sheet("Kalender Posting")

    # Styling
    week_fill = PatternFill(start_color="DDEBF7", end_color="DDEBF7", fill_type="solid")
    day_fill = PatternFill(start_color="F2F2F2", end_color="F2F2F2", fill_type="solid")
    bold_font = Font(bold=True)
    thin_border = Border(left=Side(style='thin'), right=Side(style='thin'), top=Side(style='thin'), bottom=Side(style='thin'))

    # Membuat struktur kalender
    calendar_data = {
        "Minggu 1": ["Belajar Python dari NOL", "Setup Awal JavaScript"],
        "Minggu 2": ["Perintah Dasar GIT", "10 Rumus Excel Paling Berguna"],
        "Minggu 3": ["Mempercepat Kinerja Windows", "Website Portofolio HTML & CSS"],
        "Minggu 4": ["5 Ekstensi Chrome Produktif", "Menguasai Pivot Table Excel"],
    }

    ws2.cell(row=2, column=2, value="Contoh Jadwal Posting Bulan Pertama").font = Font(bold=True, size=14)

    current_col = 2
    for week, videos in calendar_data.items():
        # Header Minggu
        ws2.merge_cells(start_row=4, start_column=current_col, end_row=4, end_column=current_col + 1)
        cell = ws2.cell(row=4, column=current_col, value=week)
        cell.fill = week_fill
        cell.font = bold_font
        cell.alignment = Alignment(horizontal='center')

        # Hari dan Konten
        ws2.cell(row=5, column=current_col, value="Rabu").fill = day_fill
        ws2.cell(row=5, column=current_col+1, value="Sabtu").fill = day_fill
        ws2.cell(row=6, column=current_col, value=videos[0]).alignment = Alignment(wrap_text=True)
        ws2.cell(row=6, column=current_col+1, value=videos[1]).alignment = Alignment(wrap_text=True)

        current_col += 3 # Beri spasi antar minggu

    # Strategi
    ws2.merge_cells("B10:J15")
    strategy_text = (
        "Strategi Jadwal Upload:\n\n"
        "• Frekuensi: 2 video per minggu untuk membangun konsistensi.\n"
        "• Waktu Upload Optimal: Rabu & Sabtu sore (sekitar 16.00 - 18.00 WIB) untuk menangkap audiens setelah jam kerja/sekolah.\n"
        "• Distribusi Konten: Seimbangkan antara konten Pemula yang menarik audiens baru dan topik tren untuk menjaga engagement."
    )
    cell = ws2.cell(row=10, column=2, value=strategy_text)
    cell.alignment = Alignment(vertical='top', wrap_text=True)
    cell.font = Font(size=12)

    # === TAB 3: ANALYTICS & PELACAKAN KINERJA ===
    ws3 = wb.create_sheet("Analytics & Target")

    # Bagian A: Target Bulanan
    ws3.cell(row=2, column=2, value="Bagian A: Pelacakan Target Bulanan").font = Font(bold=True, size=14)
    headers_ws3a = ["Bulan", "Target Subs", "Realisasi Subs", "Target Views", "Realisasi Views", "Target Engagement Rate (%)", "Realisasi Engagement Rate (%)", "Catatan & Strategi"]
    data_ws3a = [
        ("Bulan 1", 300, "", 15000, "", 15, "", "Fokus pada CTA di akhir video"),
        ("Bulan 2", 500, "", 25000, "", 15, "", "Promosi di grup Facebook & medsos"),
        ("Bulan 3", 1200, "", 50000, "", 15, "", "Mulai riset untuk kolaborasi"),
        ("Bulan 4", 1800, "", 70000, "", 15, "", None),
    ]

    ws3.append([]) # Baris kosong
    ws3.append([]) # Baris kosong
    ws3.append(headers_ws3a)
    for cell in ws3[4]:
        cell.font = header_font
        cell.fill = header_fill

    for row in data_ws3a:
        ws3.append(row)
    ws3.column_dimensions['H'].width = 40

    # Bagian B: Kinerja Video
    ws3.cell(row=11, column=2, value="Bagian B: Kinerja Video Individual").font = Font(bold=True, size=14)
    headers_ws3b = ["Judul Video", "Tgl Upload", "Views", "Likes", "Comments", "Shares", "Subscriber Baru", "Subscriber Kembali", "Growth Rate (%)", "Content Performance Index (CPI)"]
    ws3.append([]) # Baris kosong
    ws3.append([]) # Baris kosong
    ws3.append(headers_ws3b)

    start_row_b = ws3.max_row
    for cell in ws3[start_row_b]:
        cell.font = header_font
        cell.fill = header_fill

    # Tambah 10 baris kosong dengan formula
    for i in range(start_row_b + 1, start_row_b + 11):
        ws3.append(["" for _ in headers_ws3b]) # Buat baris kosong
        # Formula Growth Rate di kolom I
        growth_cell = ws3.cell(row=i, column=9)
        growth_cell.value = f'=IF(C{i}>0, ((G{i}+H{i})/C{i})*100, 0)'
        growth_cell.number_format = '0.00"%"'
        # Formula CPI di kolom J
        cpi_cell = ws3.cell(row=i, column=10)
        cpi_cell.value = f'=(C{i}*0.4)+(D{i}*10)+(E{i}*20)+(F{i}*30)'
        cpi_cell.number_format = '#,##0'

    # Atur lebar kolom untuk Bagian B
    ws3.column_dimensions['A'].width = 45
    ws3.column_dimensions['I'].width = 18
    ws3.column_dimensions['J'].width = 25

    # === TAB 4: ROADMAP & STRATEGI PENGEMBANGAN ===
    ws4 = wb.create_sheet("Roadmap Pengembangan")

    headers_ws4 = ["Fase", "Durasi", "Fokus Utama", "Jumlah Konten", "Target Subscriber", "Aktivitas Kunci"]
    data_ws4 = [
        ("1. Foundation", "Bulan 1-2", "Konten Level Pemula", "8 - 16 video", "300 - 500", "• Membangun konsistensi upload (2x seminggu)\n• Memperkuat branding channel (banner, logo)\n• Riset kata kunci dasar"),
        ("2. Growth", "Bulan 3-4", "Konten Level Menengah", "10 - 20 video", "500 - 1.000", "• Mulai berinteraksi aktif di kolom komentar\n• Promosi silang di media sosial\n• Mulai mencari peluang kolaborasi kecil"),
        ("3. Authority", "Bulan 5-6", "Konten Level Lanjut", "6 - 12 video", "1.000+", "• Mengadakan sesi Live Q&A atau webinar\n• Membuat seri konten yang mendalam\n• Membangun komunitas (misal: grup Discord/Telegram)"),
    ]

    ws4.append(headers_ws4)
    for cell in ws4[1]:
        cell.font = header_font
        cell.fill = header_fill
        cell.alignment = Alignment(horizontal='center', vertical='center')

    for row_data in data_ws4:
        ws4.append(row_data)

    for row in ws4.iter_rows(min_row=2, max_row=ws4.max_row):
        for cell in row:
            cell.alignment = Alignment(wrap_text=True, vertical='top')

    # Atur lebar dan tinggi baris
    ws4.column_dimensions['A'].width = 15
    ws4.column_dimensions['B'].width = 12
    ws4.column_dimensions['C'].width = 25
    ws4.column_dimensions['D'].width = 15
    ws4.column_dimensions['E'].width = 20
    ws4.column_dimensions['F'].width = 50
    for i in range(2, ws4.max_row + 1):
        ws4.row_dimensions[i].height = 60

    # Atur sheet aktif
    wb.active = ws1

    # Simpan file
    wb.save(filename)
    print(f"File Excel '{filename}' berhasil dibuat!")

# Jalankan fungsi untuk membuat file
create_youtube_strategy_spreadsheet()
```
